```python
"""
    DocString:
    
    Introduction:
    This game is based on the movie Catch Me If You Can (2002). 
    The story develops in three stages - hero indtroduces himself and learn to manipulate other people, 
    starts making money and finally escapes FBI. The winning prize is an offer to work at FBI Financial Crimes
    Unit. There is a function to fail as well.
    The game requires some knowledge from the movie, but also a common sense.
    
    Round 1: Learn the art of manipulation
    Round 2: Start making money
    Round 3: Escape FBI
    
"""
#Importing the library to choose a random value and to exit
from sys import exit
import random

# Defining a fucntion to initiate the game
def start():
    print("""
    ----------------------------------------------------------------------------------
     As you start on your journey, let's first develop a fun little story about you
     through the mad lib game to get to know you more. Your name is Frank
     Abagnale and you are a teenager. Let's find out more about you.
    ----------------------------------------------------------------------------------
         """)
    # Developing a mad lib game to know more about the user 
    city = input(prompt = "Please insert your city here: \n")
    hobby = input(prompt = "Please insert your hobby here: \n")
    book = input(prompt = "Please insert your favorite book here: \n")
    character = input(prompt = "Please describe your character in one word here: \n")
    dream = input(prompt = "Please insert your dream here: \n")
    
    #Asking for input from the user
    input(prompt= "<Press enter to continue>")
    
    # Printing the result of the Mad Lib game
    print(f""" Here is your story.
    ---------------------------------------------------------------------
    I currently live in {city} with my lovely parents.
    In my free time I enjoy {hobby}. My favorite book is {book}. 
    I would describe myself in one word as {character}. My dream is {dream}.
    I am excited to embark on the this journey!
    ---------------------------------------------------------------------""")
    
    # Asking for input to continue to the next stage
    input(prompt = "<Press enter to continue and let's learn the art of manipulation. >\n")
    
    # Printing the lines to vizually separate the code
    print(10*'-')
    
    # Run the next function to proceed in the game
    manipulate()
    
# Defining a necklace trick to charm women
def necklace_trick():
    # Defining a variable with user input to use later in the game
    choice2 = input(prompt = """
    Frank, this is the most important lesson you 
    will ever get to succeed in your journey.
    Please press Y to learn or N to skip.""")
            
    # Conditional IF statement
    if choice2.lower() == 'y' or choice2.lower() == 'yes':
        print("""Here it is:
        --------------------------------------------------------------------
         "Is this your necklace? I just found it in the parking lot, it must 
          have slipped right off your neck :)"
        -------------------------------------------------------------------
         This will be very helpfull throughout your criminal journey.      
               """)
        input(prompt = "<Press enter to continue>")
        make_money()

    # Conditional ELSE statement 
    elif choice2.lower() == 'n' or choice2.lower() == 'no':
        print('No probem, we will proceed further without it, Frank')
        make_money()
        
    # Conditional ELSE statment        
    else:
        print("Something went wrong.")
        necklace_trick()
            
        
# Defining a new fuction to proceed to the next stage
def manipulate():
    
    # Printing the instructions to the challenge in the stage
    print("""
    --------------------------------------------------------------------------------------
    Try to guess who was your inspiration to learn and apply the art of manipulating people:
    1) Robert Green, author of 48 Laws of Power and Art of Seduction
    2) Your father, Frank Abagnae Sr.
    3) Nobody, I just want to quit this silly game.
    --------------------------------------------------------------------------------------
    Please select the number of the answer.
    """)
        
    # Requesting input from the user to advance the game
    choice = input(prompt = ">  \n")
    
    # Conditional If statement for the first option
    if '1' in choice or 'one' in choice.lower() or 'green' in choice.lower() or 'book' in choice.lower():
        print("""
    ----------------------------------------------------------------------
    I know you like reading Frank, but this is incorrect. You have
    one more chance.
    ----------------------------------------------------------------------
        """ )
        # While loop to count chances
        chance = 1
        while chance > 0:
            manipulate()
            chance -= 1
            
            #conditional statements to fail or continue the game
            if chance == 0:
                fail()
            else:
                continue 
    
    # Conditional ELIF statement for the second option
    elif '2' in choice or 'two' in choice.lower() or 'father' in choice.lower():
        print("""
    ----------------------------------------------------------------------
    This is correct, Frank. Your father is a great inspiration to you.
    ----------------------------------------------------------------------
        """)
        
        # Requesting an input from the user
        input(prompt = """<Press enter to continue and let's learn the famous necklace 
              trick you should learn. >\n""")
        necklace_trick()

    # Conditional ELIF statement for the third condition
    elif '3' in choice or 'three' in choice.lower() or 'quit' in choice.lower:
        print('You have all the rights, Frank.')
        exit(0)
        
# Defining a new function to proceed to the next level
def make_money():
    print("""
    --------------------------------------------------------------------------
    Frank, you now know how to use the art of manipulation on people.
    Time to make money and help your old man out.
    Please choose the ultimate fraudulent career path for yourself.
    
    1) Fake Co-pilot and Conterfeiter.
    2) Fake Python Developer.
    3) I just want to quit.
    --------------------------------------------------------------------------
    """)
    # Requesting an input from the user to get an answer
    answer = input(prompt = '')
    
    # Conditional statements to process the input from the user
    
    if '1' in answer or 'one' in answer.lower() or 'co-pilot' in answer.lower() or 'conterfeiter' in answer.lower():
        print("That's correct, you get to proceed further")
        escape_fbi()
            
    elif '2' in answer or 'two' in answer.lower() or 'python' in answer.lower():
        choice = 1
        print("You can never fake a python code. Try again.")
        
        #while loop to count the chances
        while choice > 0:
            make_money()
            choice -= 1
            if choice == 0:
                fail()
        
    elif '3' in answer or 'three' in answer.lower() or 'quit' in answer.lower():
        input(promt = '<Press enter to exit>/n')
        exit(0)
        
    else:
        # while loop to count the chances
        rounds = 2
        while rounds > 0:
            print("""
              Something went wrong, you must enter 1,2 or 3.
              Please try again or you will fail""")
            make_money()
            
            rounds -= 1
            
            # Conditional statements to fail or continue the game
            if rounds == 0:
                fail()
            else:
                continue
            
# Defining a new function to proceed to a next stage
def escape_fbi():
    print("""-----------------------------------------------------------------
    Frank, you managed to make a fuss in the FBI. 
    They can try to catch you if they can. The question is - can they?""")

    input(prompt = '<Press enter to continue>')
    
    # Printing instructions for the next stage
    print("""
    --------------------------------------------------------------------------
    You are in a hotel. Your room is full of checks and stamps for your counterfeit
    activities. FBI agent Carl Hanratty walks into your hotel room. There he is,
    standing before you. You are sweating, prison cell flashes before your eyes.
    What do you do? 
    1) Pretend to be a Secret Services Agent, Barry Allen.
    2) Kill the guy
    3) Give him a lot of money to let you go.
    --------------------------------------------------------------------------
    """)
    
    # Requesting an input from the user to get an answer
    answer = input(prompt = "Please choose the number: ")
    
    # IF conditional statement for the different options
    if '1' in answer or 'one' in answer.lower() or "pretend" in answer.lower() or 'secret services agent' in answer.lower():
        print('That is correct, you are brilliant.')
        input(prompt = '<Press enter to continue>.')
        
        # Creating a new variable inside an IF conditional to make a nested conditional for additional steps
        barry_allen = input(prompt = """By the way, who is Barry Allen? If you know, you are great, 
            if you don't know, type "tell me": """)
        
        # Try/Except clase to catch the errors
        try:
            if barry_allen.lower() == 'tell me':
                print("He's THE FLASH. You are just a teenager who loves comic books.")
            
            else: 
                print('Something went wrong, we will proceed.')
        except:
            print("Something went wrong. Barry Allen is THE FLASH, let's continue")
        
        # Requesting an input from the user to continue
        input(prompt = '<Press enter to continue>')
        win()
        
    # ELIF conditional statement for the answer to the main question        
    else:
        another_chance = input(prompt = """Frank, it's been a great game, but you lost. You can have another chance,
              do you want to use it? Type 'yes'. """)
        
        # Nested conditional to give a second chance
        if 'y' in another_chance.lower() or 'yes' in another_chance.lower():
            escape_fbi()
        else:
            fail()
        
        # Printing the lines to vizually separate the code
        print(10*'-')

                
# Defining a new function to win the game        
def win():
    print("""
    --------------------------------------------------------------------------
    Frank, you won. You are now a bestselling author, you work for 
    FBI instead of going to prison and Leonardo DiCaprio will be playing you in a movie.
    --------------------------------------------------------------------------
    """)
    exit(0)

# Defining a new function to fail the game  
def fail():
    print('\nYou have failed in your quest to becoming the youngest fraud in history.')
    input(prompt = '<Game Over>')
    exit(0)
    
    
start()
    
    
```
